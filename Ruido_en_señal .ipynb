```python
import numpy as np
import matplotlib.pyplot as plt

# Generar señal
t = np.linspace(0, 10, 1000)
x = np.sin(2 * np.pi * t)

# Generar ruido
ruido = 0.1 * np.random.randn(len(t))

# Sumar ruido a la señal
y = x + ruido

# Graficar señal original y ruidosa
plt.plot(t, x, label='Señal original')
plt.plot(t, y, label='Señal con ruido')
plt.legend()
plt.show()
```

En este ejemplo, primero se genera una señal senoidal de 1 kHz. Luego, se genera un ruido aleatorio, que genera valores aleatorios con una distribución normal estándar (media cero y desviación estándar uno) multiplicados por 0.1 para ajustar la amplitud del ruido.
