```python
import numpy as np
import matplotlib.pyplot as plt
fs = 1000
# Definir la señal y la frecuencia de muestreo
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2*np.pi*10*t) + 0.5*np.sin(2*np.pi*20*t)


# Calcular la energía y la potencia de la señal
energy = np.sum(np.square(x))
power = np.mean(np.square(x))

# Graficar la señal
plt.figure()
plt.plot(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal')

# Mostrar la energía y la potencia de la señal
plt.text(0.05, 0.95, f'Energía = {energy:.2f}', transform=plt.gca().transAxes)
plt.text(0.05, 0.9, f'Potencia = {power:.2f}', transform=plt.gca().transAxes)

plt.show()
```
