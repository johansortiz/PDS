```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

b = [1, 1] # Coeficientes del numerador
a = [1, -0.5] # Coeficientes del denominador
r, p, k = signal.residue(b, a) # Calcula los coeficientes de fracciones parciales

# Calcula la transformada Z inversa
n = np.arange(0, 10)
y = np.zeros(len(n))
for i in range(len(r)):
    y += r[i] * p[i]**n

# Grafica la respuesta
plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Transformada Z inversa')
plt.show()
```
