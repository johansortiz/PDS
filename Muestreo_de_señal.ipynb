```python
import numpy as np
import matplotlib.pyplot as plt

# Generar señal original
f = 400 # Frecuencia de la señal
fs = 2.5*f # Frecuencia de muestreo
t = np.arange(0, 2*np.pi, 1/fs)
senal = np.sin(2 * np.pi * t)

# Graficar espectros de frecuencia

plt.figure()
plt.plot(t, senal)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
