```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la ecuaci칩n de diferencia del sistema
a = [1, -1.5, 0.7]
b = [1, 0.5, 0.2]
system = (b, a)

# Definir una se침al de entrada
x = np.sin(np.linspace(0, 10*np.pi, 100))

# Aplicar la se침al de entrada al sistema
y1 = np.convolve(x, b, mode='same')
y2 = np.convolve(x, a, mode='same')

# Graficar la se침al de entrada y de salida
plt.figure()
plt.plot(x, label='Entrada')
plt.plot(y1, label='Salida 1')
plt.plot(y2, label='Salida 2')
plt.xlabel('Tiempo (muestras)')
plt.ylabel('Amplitud')
plt.title('Linealidad e invarianza temporal')
plt.legend()
plt.show()
```
