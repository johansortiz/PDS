```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Generar señal digital
fs = 1000 # Frecuencia de muestreo
f = 10 # Frecuencia de la señal
t = np.linspace(0, 1, fs)
x = np.sin(2 * np.pi * f * t)

# Aplicar filtro paso bajo
fc = 50 # Frecuencia de corte del filtro
b, a = signal.butter(4, fc/(fs/2), 'low')
y = signal.lfilter(b, a, x)

# Graficar señal digital y analógica
plt.plot(t, x, label='Señal digital')
plt.plot(t, y, label='Señal analógica')
plt.legend()
plt.show()
```

En este ejemplo, primero se genera una señal senoidal de 10 Hz y se muestrea a una frecuencia de 1000 Hz
Luego, se aplica un filtro paso bajo para generar los coeficientes del filtro. El filtro paso bajo se define con una frecuencia de corte de 50 Hz, lo que significa que atenúa las frecuencias por encima de 50 Hz y deja pasar las frecuencias por debajo de ese valor.
