```python
import numpy as np
fs = 1000
# Definir la señal y la frecuencia de muestreo
t = np.linspace(0, 1, fs, endpoint=False)
x = np.sin(2*np.pi*10*t)


# Calcular la diferencia hacia atrás de la señal
diff_x = np.diff(x, prepend=x[0])

# diff_x contiene la diferencia hacia atrás de la señal "x"
print(diff_x)
```
