```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import zpk2tf, freqz

# Definir los coeficientes del polinomio del numerador, denominador y los ceros y polos
num = [1, 0, 0, 0]
den = [1, -1.8*np.exp(-1j*np.pi/4), 0.81*np.exp(-1j*np.pi/2)]
z = [0, 0]
p = [0.9*np.exp(1j*np.pi/4), 0.9*np.exp(-1j*np.pi/4)]

# Convertir los coeficientes a la representación de polinomios
b, a = zpk2tf(z, p, 1)

# Calcular la respuesta en frecuencia
w, h = freqz(b, a)

# Graficar la respuesta en frecuencia en dB
plt.plot(w/np.pi, 20*np.log10(np.abs(h)))
plt.title('Respuesta en frecuencia de la transformada Z')
plt.xlabel('Frecuencia normalizada')
plt.ylabel('Magnitud (dB)')
plt.show()
```
