```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Definir los coeficientes de la ecuación de diferencia
b = [2, -3, 1]
a = 1

# Generar una señal de entrada
x = np.sin(np.linspace(0, 10*np.pi, 100))

# Aplicar el filtro al sistema
y = signal.lfilter(b, a, x)

# Graficar la señal de entrada y de salida
plt.plot(x, label='Entrada')
plt.plot(y, label='Salida')
plt.xlabel('Tiempo (muestras)')
plt.ylabel('Amplitud')
plt.title('Sistema no recursivo')
plt.legend()
plt.show()
```
