```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generar se√±al discreta
fs = 200 # Frecuencia de muestreo
t = np.arange(0, 1, 1/fs) # Vector de tiempo
x = np.sin(2*np.pi*t) # Se√±al senoidal

# Crear DataFrame con se√±al discreta
df = pd.DataFrame({'tiempo': t, 'se√±al': x})

# Visualizar DataFrame
print(df.head(10)) # Imprimir primeros 10 valores de la se√±al

# Graficar se√±al discreta
plt.stem(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

Representacion en secuencia

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir funci√≥n para generar se√±al discreta
def signal_func(n):
    return np.sin(2*np.pi*0.1*n) + 0.5*np.sin(2*np.pi*0.3*n)

# Generar se√±al discreta
n = np.arange(0, 80)
x = signal_func(n)

# Graficar se√±al discreta
plt.stem(n, x, use_line_collection=True)
plt.xlabel('Tiempo discreto (n)')
plt.ylabel('Amplitud')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar funci√≥n de impulso discreto
n = np.arange(-10, 11)
impulse = np.zeros(len(n))
impulse[len(n)//2] = 1

# Graficar funci√≥n de impulso discreto
plt.title('Impulso unitario')
plt.stem(n, impulse, use_line_collection=True)
plt.xlabel('Tiempo discreto (n)')
plt.ylabel('Amplitud')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir par√°metros de la se√±al exponencial
A = 1   # amplitud
tau = 10    # constante de tiempo
n = np.arange(0, 50)    # vector de tiempo discreto

# Generar se√±al exponencial decreciente
x = A * np.exp(-n/tau)

# Graficar se√±al exponencial decreciente
plt.title('Exponencial decreciente')
plt.plot(n, x)
plt.xlabel('Tiempo discreto (n)')
plt.ylabel('Amplitud')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir par√°metros de la se√±al sinusoidal
A = 1   # amplitud
f = 1   # frecuencia
phi = np.pi/2  # fase

# Definir funci√≥n sinusoidal
def sinusoidal(x):
    return A * np.sin(2*np.pi*f*x + phi)

# Generar valores de x
x = np.linspace(0, 2*np.pi, 100)

# Evaluar la funci√≥n sinusoidal en los valores de x
y = sinusoidal(x)

# Graficar la funci√≥n sinusoidal
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funci√≥n sinusoidal')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n rampa
def rampa(x):
    return np.maximum(x, 0)

# Generar valores de x
x = np.linspace(-5, 5, 100)

# Evaluar la funci√≥n rampa en los valores de x
y = rampa(x)

# Graficar la funci√≥n rampa
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funci√≥n rampa')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n exponencial compleja
def exponencial_compleja(x):
    return np.exp(1j * x)

# Generar valores de x
x = np.linspace(-np.pi, np.pi, 100)

# Evaluar la funci√≥n exponencial compleja en los valores de x
y = exponencial_compleja(x)

# Graficar la funci√≥n exponencial compleja
plt.plot(x, np.real(y), label='Parte real')
plt.plot(x, np.imag(y), label='Parte imaginaria')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funci√≥n exponencial compleja')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la se√±al peri√≥dica
def signal_periodic(x):
    return np.sin(2 * np.pi * x)

# Definir la se√±al aperi√≥dica
def signal_aperiodic(x):
    return np.exp(-x)

# Generar valores de x
x = np.linspace(0, 10, 1000)

# Evaluar las se√±ales en los valores de x
y_periodic = signal_periodic(x)
y_aperiodic = signal_aperiodic(x)

# Graficar las se√±ales
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y_periodic, label='Se√±al peri√≥dica')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Se√±al peri√≥dica')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y_aperiodic, label='Se√±al aperi√≥dica')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Se√±al aperi√≥dica')
plt.legend()

plt.show()
```
