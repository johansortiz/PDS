```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos una señal analógica de ejemplo
t = np.linspace(0, 1, 100)    # Vector de tiempo
f = 5                          # Frecuencia de la señal en Hz
x = np.sin(2*np.pi*f*t)        # Señal analógica

# Digitalizamos la señal
fs = 100                       # Frecuencia de muestreo en Hz
Ts = 1/fs                      # Período de muestreo
n = np.arange(0, len(t))      # Vector de índices de la señal
xn = np.sin(2*np.pi*f*n*Ts)    # Señal digital

# Graficamos las señales analógica y digital
plt.figure()
plt.plot(t, x, label='Señal analógica')
plt.stem(n*Ts, xn, label='Señal digital', use_line_collection=True)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
